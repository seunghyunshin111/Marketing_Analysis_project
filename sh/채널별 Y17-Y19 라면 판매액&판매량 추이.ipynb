```python
data = pd.read_excel("C:/Marketing_project/sh/라면데이터.xlsx")
```

```python
data.head()
```

```python
data = data[data["채널"]!="전국"]
data
```

```python
data['채널'].unique()
```

```python
import numpy as np

data['판매총액'] = 0
data['판매총액'] = data.loc[:, "JAN17":"DEC17"].fillna(0).replace("ERR", 0).apply(np.sum, axis=1)
data['판매총액'].head()
```

```python
data['FACT'].unique()
```

```python
# 판매액 
sales_rate_idx = data['FACT'] == '판매액'

# 판매량
sales_amount_idx = data['FACT'] == '판매량 (UNIT)'
```

```python
# 채널별 
sales_rate = data[sales_rate_idx]['판매총액'].groupby(data['채널'])
print()
print("="*10)
print("17 채널별 판매액")
print("="*10)
print(sales_rate.sum())

sales_amount = data[sales_amount_idx]['판매총액'].groupby(data['채널'])
print()
print("="*10)
print("17 채널별 판매량")
print("="*10)
print(sales_amount.sum())
```

```python
sales_sum_17 = sales_rate.sum()
sales_amount_17 = sales_amount.sum()
sales_amount_17
```
